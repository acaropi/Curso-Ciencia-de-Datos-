**P4.** Crea una función llamada `calcular_area` que reciba `base` y `altura` y devuelva el área de un triángulo (`base * altura / 2`).  
Llama la función con `base=5` y `altura=8` e imprime el resultado.

```python
# Tu respuesta aquí def (para definir la función)

#CALCULA
def calcular_area(base, altura):  # (base, altura) dato que entra
  return base*altura/2  #Dato que sale

area = calcular_area(5,8)  #Especifico los varores de base y altura
print(area)  #Imprimo el resultado
```

```python
#Realizo otro ejemplo de función - Covertir de °C a F -  Fórmula: F=(C×9/5)+32

#CONVIERTE
def celsius_a_fahrenheit(celsius):
  return (celsius*9/5)+32

resultado = celsius_a_fahrenheit(25)
print(resultado)

#DECIDE
def es_mayor_de_edad(edad):
    if edad >= 18:
        return "Es mayor de edad"
    else:
        return "Es menor de edad"

print(es_mayor_de_edad(20))
print(es_mayor_de_edad(15))
```

**P5.** Usa una **list comprehension** para crear una lista con los cuadrados de los números del 1 al 10.  
Imprime la lista resultante.

```python
# Tu respuesta aquí # Una "list comprehension" es un for más corto
# El cuadrado de un número es (n * n) o (n^2) que este símbolo para elevar "^" en Python es (n**n)

cuadrados = [n**2 for n in range(1, 11)]  # Para cada n del rango del 1 al 10 guarda su cuadrado en una lista
print(cuadrados)
```

**P6.** Crea los siguientes arreglos e imprime su `shape` y `dtype`:  
- Un arreglo con los valores `[1.5, 2.5, 3.5, 4.5]`  
- Una matriz de ceros de 3 filas × 4 columnas  
- Un arreglo con 6 valores igualmente espaciados entre 0 y 1 (usando `linspace`)

```python
# Tu respuesta aquí
#Arreglo con los valores

arr1 = np.array([1.5, 2.5, 3.5, 4.5])
print(arr1)
print("shape",arr1.shape)
print("shape",arr1.dtype)

#Matriz 3 filas × 4 columnas
# Para matrices de ceros se usa np.zeros((filas, columnas)).

matriz_ceros = np.zeros((3,4))
print(matriz_ceros)

print("Shape:", matriz_ceros.shape)
print("Dtype:", matriz_ceros.dtype)

#arreglo con 6 valores igualmente espaciados entre 0 y 1 (usando linspace)
#np.linspace(inicio, fin, cantidad)

arr_linspace = np.linspace(0,1,6)

print(arr_linspace)
print("Shape:", arr_linspace.shape)
```

**P7.** Dado el arreglo `a = np.arange(1, 13)`, usa `reshape` para convertirlo en una matriz de 3 × 4.  
Luego imprime el valor máximo, el valor mínimo y la posición (índice) del máximo.

```python
a = np.arange(1, 13)
print(a)

# Tu respuesta aquí

# Convertir el arreglo en una matriz de 3 × 4 con reshape

matriz = a.reshape(3, 4)
print("------------------------------")
print(matriz)

#Imprimir el valor máximo y mínimo
print("------------------------------")
print("Valor máximo:", matriz.max())
print("Valor mínimo:", matriz.min())

#Obtener la posición (índice) del valor máximo

fila, columna = np.unravel_index(matriz.argmax(), matriz.shape)
print("------------------------------")
print("Posición del valor máximo → fila:", fila, ", columna:", columna)
```
